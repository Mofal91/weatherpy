```python
import matplotlib.pyplot as plt
import requests
from random import uniform
from citipy import citipy
import pandas as pd
import numpy as np
```

```python
# numpy.random.uniform(low=0.0, high=1.0, size=None)
humidity=[]
wind=[]
cloudy=[]
date=[]
temp=[]

lat = np.random.uniform(low=-90.0, high=90.0,size=500)
long = np.random.uniform(low=-180.0, high=180.0,size=500)

api_key="ad777218ae52dfd2085f8a2b1d497e3f"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

query_url = f"{url}appid={api_key}&units={units}&q="

cities=[citipy.nearest_city(lat[i],long[i]).city_name for i in range(0,500)]
city=cities
#feedback=[requests.get(query_url+city).json() for city in cities]
#response = feedback

weather_df=pd.DataFrame({"latitude":lat,"Longitude":long,"cities":city})
weather_df.head()
```

```python
for index,rows in weather_df.iterrows():
    response=requests.get(query_url+city[index]).json()
    try:
        temp=response["main"]["temp"]
        humidity=response["main"]["humidity"]
        wind=response["wind"]["speed"]
        cloudy=response["clouds"]["all"]
        
        weather_df.set_value(index, "Max Temperature (F)", pd.to_numeric(temp))
        weather_df.set_value(index, "Humidity (%)", humidity)
        weather_df.set_value(index, "Cloudiness (%)", cloudy)
        weather_df.set_value(index, "Wind Speed (mph)", wind)
    except:
        print("error")
```
